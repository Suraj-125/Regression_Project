```python
import numpy as np
import pandas as pd

import warnings
warnings.simplefilter("ignore")
```

```python
df = pd.read_csv("Laptop_analyzed.csv")
df.drop("Unnamed: 0", axis = 1, inplace = True)
df.head()
```

```python
X = df.drop("Price", axis = 1)
y = df["Price"]
```

```python
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.15, random_state = 72)
```

# ***Perceptron***

```python
def perceptron():
    model = Sequential()
    model.add(Dense(input_dim = 41, units = 81, activation = "relu", kernel_initializer = "uniform"))
    model.add(Dense(units = 1, activation = "relu", kernel_initializer = "uniform"))
    model.compile(optimizer = "adam", loss = "mean_squared_error")
    return model
```

```python
process = ColumnTransformer(transformers = [
    ("encode", OneHotEncoder(sparse = False, drop = "first"), [0, 1, 4, 10, 14]),
    ("scale", StandardScaler(), [2, 3, 5, 6, 11, 12, 13])
], remainder = "passthrough")

X_train = process.fit_transform(X_train)
X_test = process.transform(X_test)
```

```python
model = perceptron()
model.fit(X_train, y_train, epochs = 100, batch_size = 32)
```

```python
pred_train = model.predict(X_train)
pred_test = model.predict(X_test)
print("Training Accuracy :", r2_score(y_train, pred_train))
print("Testing Accuracy :", r2_score(y_test, pred_test))
print("MAE :", mean_absolute_error(y_test, pred_test))
print("MSE :", mean_squared_error(y_test, pred_test))
```

# ***ANN with 1 Hidden Layer***

```python
def ann_1_hidden_layer():
    model = Sequential()
    model.add(Dense(input_dim = 41, units = 81, activation = "relu", kernel_initializer = "uniform"))
    model.add(Dense(units = 81, activation = "relu", kernel_initializer = "uniform"))
    model.add(Dense(units = 1, activation = "relu", kernel_initializer = "uniform"))
    model.compile(optimizer = "adam", loss = "mean_squared_error")
    return model
```

```python
model_ann1 = ann_1_hidden_layer()
model_ann1.fit(X_train, y_train, epochs = 100, batch_size = 32)
```

```python
pred_train = model_ann1.predict(X_train)
pred_test = model_ann1.predict(X_test)
print("Training Accuracy :", r2_score(y_train, pred_train))
print("Testing Accuracy :", r2_score(y_test, pred_test))
print("MAE :", mean_absolute_error(y_test, pred_test))
print("MSE :", mean_squared_error(y_test, pred_test))
```

# ***ANN with 2 hidden layer***

```python
def ann_2_hidden_layer():
    model = Sequential()
    model.add(Dense(input_dim = 41, units = 81, activation = "relu", kernel_initializer = "uniform"))
    model.add(Dense(units = 81, activation = "relu", kernel_initializer = "uniform"))
    model.add(Dense(units = 81, activation = "relu", kernel_initializer = "uniform"))
    model.add(Dense(units = 1, activation = "relu", kernel_initializer = "uniform"))
    model.compile(optimizer = "adam", loss = "mean_squared_error")
    return model
```

```python
model_ann2 = ann_2_hidden_layer()
model_ann2.fit(X_train, y_train, epochs = 100, batch_size = 32)
```

```python
pred_train = model_ann2.predict(X_train)
pred_test = model_ann2.predict(X_test)
print("Training Accuracy :", r2_score(y_train, pred_train))
print("Testing Accuracy :", r2_score(y_test, pred_test))
print("MAE :", mean_absolute_error(y_test, pred_test))
print("MSE :", mean_squared_error(y_test, pred_test))
```

```python
# It is overfitting . Now we reduce the neurons .

def ann_2_hidden_layer():
    model = Sequential()
    model.add(Dense(input_dim = 41, units = 81, activation = "relu", kernel_initializer = "uniform"))
    model.add(Dense(units = 54, activation = "relu", kernel_initializer = "uniform"))
    # model.add(Dense(units = 54, activation = "relu", kernel_initializer = "uniform"))
    model.add(Dense(units = 1, activation = "relu", kernel_initializer = "uniform"))
    model.compile(optimizer = "adam", loss = "mean_squared_error")
    return model
```

```python
model_ann2 = ann_2_hidden_layer()
model_ann2.fit(X_train, y_train, epochs = 100, batch_size = 32)
```

```python
pred_train = model_ann2.predict(X_train)
pred_test = model_ann2.predict(X_test)
print("Training Accuracy :", r2_score(y_train, pred_train))
print("Testing Accuracy :", r2_score(y_test, pred_test))
print("MAE :", mean_absolute_error(y_test, pred_test))
print("MSE :", mean_squared_error(y_test, pred_test))
```

***Moving with 1 hidden layer and 54 neurons generalize the ann model . Since, accuracy is not too as like SVR. Hence we consider SVR as final model***
